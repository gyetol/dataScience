#### 실습02

```python

def calc(operator,*args):
   
    if operator == "+":
        ret = 0
        for arg in args: 
                ret +=arg
        print(ret)
    elif operator == "*":
        ret = 1
        for arg in args: 
                ret *=arg
        print(ret)
    else:
            print("아직 구현되지 않은 기능입니다.")
            
calc("+",1,2,3,4,5)
calc("*",1,2,3,4,5)
```

## 재귀함수

```python
def factorial(n):
    if n == 1:
        return 1
    return n * factorial(n-1)

factorial(5)
```

#### 실습03

```python
def f_sum(arg):
    if arg==1:
        return 1
    return arg+ f_sum(arg-1)

f_sum(5)
```

```python
def f_number(arg):
    if (int(arg/10))==0:
        print(arg)
    f_number(int(arg/10))
    print()
f_number(1234)
```

##### 은이님 코드

```python
def f_number(n):
    if n==0:
        return
    f_number(n//10) # or f_number(int(n/10))
    print(n%10)

f_number(12445)
```
