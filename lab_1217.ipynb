```python
def tellAges(age):
    if age <10:
        return '유아'
    elif age <20:
        return '10대'
    elif age<30:
        return '20대'
    elif age<40:
        return '30대'
    elif age<50:
        return '40대'
    elif age<60:
        return '50대'
    elif age<70:
        return '60대'
    else:
        return '노인'
```

```python
titan['Range'] = titan['Age'].apply(tellAges)
```

```python
titan
```

```python
titan.groupby('Range').sum()['Survived']
```

```python
titan.groupby('Range')['Survived'].sum()  # sum()한 것에서 또 sum()을 하면 총 합계를 구할 수 있다.
```
